```python
import time
from kafka import KafkaConsumer
import json
from kafka import KafkaProducer
from afinn import Afinn
from pattern.fr import sentiment
```

```python
topic_name = "raw-tweets"
topic_en = "en-tweets"
topic_fr = "fr-tweets"
topic_n = "negative-tweets"
topic_p = "positive-tweets"

from kafka import KafkaConsumer
import json

consumer_archive = KafkaConsumer(topic_name,topic_en , topic_fr, topic_n, topic_p, value_deserializer=lambda m: json.loads(m.decode('utf-8')))

with open("archive.txt", "a") as f:
    for message in consumer_archive:
        try:
            print(message.value)
            f.write(str(message.value) + "\n")
        except Exception as e:
            print(e)
```

```python
topic_n = "negative-tweets"
consumer_n = KafkaConsumer(topic_n, bootstrap_servers="localhost:9092",value_deserializer=lambda x: json.loads(x.decode('utf-8')))

for msg in consumer_n:
    msg = json.loads(msg.value)
    print(msg)

    
```
